```python
df['bathrooms'] = df['bathrooms'].astype('category')
df['bedrooms'] = df['bedrooms'].astype('category')
df['beds'] = df['beds'].astype('category')
df['security_deposit'] = df['security_deposit'].astype('category')
df['cleaning_fee'] = df['cleaning_fee'].astype('category')
df['guests_included'] = df['guests_included'].astype('category')
df['extra_people'] = df['extra_people'].astype('category')
df['minimum_nights'] = df['minimum_nights'].astype('category')
df['maximum_nights'] = df['maximum_nights'].astype('category')
df['number_of_reviews'] = df['number_of_reviews'].astype('category')
df['cancellation_policy'] = df['cancellation_policy'].astype('category')
df['calculated_host_listings_count'] = df['calculated_host_listings_count'].astype('category')
df['accommodates.norm'] = df['accommodates.norm'].astype('category')

df['occupied_cat'] = df['occupied_cat'].astype('category').cat.codes
```

```python
varSel = pd.DataFrame({'Variable': df.columns[~df.columns.isin(['occupied_cat'])]})
varSel.head()
```

## Multivariable Analysis

```python
### drop na
df = df.dropna()
```

```python
df.head()
```

```python
## remove unnecessary vars
X = df.drop(['occupied_cat'], axis=1)
X.head()
```

```python
y = df.loc[:, 'occupied_cat']
print([X.shape,y.shape])
```

```python
varSel
```

### Variable Selection using LASSO (L1 penalization)

```python
from sklearn.linear_model import Lasso
from sklearn.feature_selection import SelectFromModel

lassomod = Lasso(alpha=0.01).fit(X, y)
```

```python
model = SelectFromModel(lassomod, prefit=True)
model.get_support()
```

```python
len(model.get_support())
varSel.shape
X.shape
```

```python
varSel['Lasso'] = model.get_support().astype('int64')
varSel
```

### Variable Selection using Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import SelectFromModel

rfmod = RandomForestClassifier().fit(X, y)
#rfmod.feature_importances_ 
```

```python
model = SelectFromModel(rfmod, prefit=True)
model.get_support()
```

```python
varSel['RandomForest'] = model.get_support().astype('int64')
varSel
```

### Variable Selection using Gradient Boosting classification

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.feature_selection import SelectFromModel

gbmod = GradientBoostingClassifier().fit(X, y)
```

```python
model = SelectFromModel(gbmod, prefit=True)
model.get_support()
```

```python
varSel['GradientBoost'] = model.get_support().astype('int64')
varSel
```

### Variable Selection using SVM classification

```python
from sklearn.svm import LinearSVC
from sklearn.feature_selection import SelectFromModel

svmmod = LinearSVC(C=0.01, penalty="l1",dual=False).fit(X, y)
```

```python
model = SelectFromModel(svmmod, prefit=True)
model.get_support()
```

```python
varSel['SVM'] = model.get_support().astype('int64')
varSel
```

### Summarization and Selection of Variables 

```python
varSel['Sum'] =  np.sum(varSel,axis=1)
varSel
```

```python
varSel.groupby('Sum')['Variable'].count()
```

We can now decide a threshold for selecting our variables!

```python
varSel[varSel.Sum > 2].Variable.values
```
